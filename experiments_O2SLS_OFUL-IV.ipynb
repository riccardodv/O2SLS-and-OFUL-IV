```python
def sherman_morrison(invA, u, v):
    d = u.shape[0]
    u = u.reshape(d,1)
    v = v.reshape(d,1)
    return  (invA - (invA @ u @ v.T @ invA)/(1 + v.T @ invA @ u))
    
def calc_sqrt_b_ridge(k,sigma_2,lamda,delta,S_hat,logdetG):
    '''Returns confidence interval width for ridge regression'''
    return sigma_2 * np.sqrt(logdetG - k * np.log(lamda) + np.log(1/(delta**2))) + np.sqrt(lamda) * S_hat 

def calc_sqrt_b_2sls(k,sigma_2,lamda,delta,S_hat,logdetG):
    '''Returns confidence interval width for 2SLS regression'''
    return sigma_2 * np.sqrt(logdetG - k * np.log(lamda) + np.log(1/(delta**2))) #+ np.sqrt(lamda) * S_hat
```

```python
def run_bandit(pmts_string, Z, X, Y, theta_star, beta_star, S_hat, sigma_1, sigma_2,lamda, lamda_IV, delta, pulled_indeces, algorithm='ridge', seed=42):
    '''function used to run an instance of linear bandits using OFUL with ridge or 2SLS regression'''
    np.random.seed(seed)
    nb_exp, T, nb_actions, dx = X.shape
    assert (nb_exp, T, nb_actions) == Z.shape[:-1]
    dz = Z.shape[-1]
    regret = np.zeros((nb_exp,T))
    pred_norms = np.zeros((nb_exp,T))
    x_inv_norms = np.zeros((nb_exp,T))
    norm_beta_hat__beta = np.zeros((nb_exp, T))
    beta_hat_rec = np.zeros((nb_exp,T,dx))
    ThetaHat_rec = np.zeros((nb_exp,T, dz, dx))
    
    print(pmts_string)
    for i in tqdm(range(nb_exp)):
        
        Ghxhx, invGhxhx = lamda*np.eye(dx), np.eye(dx)/lamda
        Gxx, invGxx = lamda*np.eye(dx), np.eye(dx)/lamda
        Gzz, invGzz = lamda*np.eye(dz), np.eye(dz)/lamda # TODO is the regularizer correct here???
        ZXt =  lamda_IV*np.eye(dz, dx) # TODO np.zeros((dz, dx))
        Xyt = np.zeros(dx) 
        hXyt = np.zeros(dx)

        logdetG = dx*np.log(lamda)
        sqrt_b_ridge = calc_sqrt_b_ridge(dx, sigma_2, lamda, delta, S_hat, logdetG)
        sqrt_b_2sls = calc_sqrt_b_2sls(dx, sigma_2, lamda, delta, S_hat, logdetG)
        
        beta_hat = np.zeros(dx)
        ThetaHat = invGzz @ ZXt
        # invThetaHat = np.eye(dx)/lamda 
        invH = np.linalg.inv(ThetaHat.T @ Gzz @ ThetaHat) #np.zeros((dx,dx)) #np.ones((dx,dx))/lamda

        for t in range(T):
            z = Z[i,t] ; x = X[i,t]; y = Y[i,t]


            # Compute INDECES
            if algorithm == 'ridge':
                # Indeces
                X_invGt_norm_sq = np.sum(np.dot(x, invGxx) * x, 1)
                preds = np.dot(x, beta_hat)
                idx = preds + sqrt_b_ridge * np.sqrt(X_invGt_norm_sq) 
                # logs for graphs 
                x_inv_norms[i,t] = max(X_invGt_norm_sq)
                norm_beta_hat__beta[i,t] = np.linalg.norm(beta_hat-beta_star)
                pred_norms[i,t] = np.linalg.norm(preds)

            if algorithm == '2sls':
                # Indeces
                X_invH_norm_sq = np.sum((x @ invH) * x,1)
                preds = np.dot(x, beta_hat)
                idx = preds + sqrt_b_2sls * np.sqrt(X_invH_norm_sq) 
                # logs for graphs 
                norm_beta_hat__beta[i,t] = np.linalg.norm(beta_hat-beta_star)
                pred_norms[i,t] = np.linalg.norm(preds)


            # Pull the arm with the highest index (OFU)
            pulled_idx = np.argmax(idx)  
            pulled_indeces[i,t] = pulled_idx 
            xt = x[pulled_idx, :]
            zt = z[pulled_idx, :]
            yt = y[pulled_idx]

            # Compute REGRET
            regret[i,t] = np.max(np.dot(x, beta_star))-np.dot(xt, beta_star) # expected regret
            # regret[i,t] = np.max(y) - yt


            # build ESTIMATORS and INDECES
            if algorithm == 'ridge':
                Xyt += xt*yt
                Gxx += np.outer(xt, xt)
                if (t % 50 == 0):
                    invGxx = np.linalg.inv(Gxx)
                else:
                    invGxx = sherman_morrison(invGxx, xt,xt) 
                logdetG += xt.T @ invGxx @ xt
                sqrt_b_ridge = calc_sqrt_b_ridge(dx, sigma_2, lamda, delta, S_hat, logdetG)
                beta_hat = invGxx @ Xyt
            
            if algorithm == '2sls':
                # Compute the predicted covariate
                hxt = ThetaHat.T @ zt
                # compute beta estimator
                hXyt += hxt*yt
                Ghxhx += np.outer(hxt, hxt) #TODO check reularizer at start
                Gzz += np.outer(zt, zt)  
                ZXt += np.outer(zt, xt)
                invH = np.linalg.inv(ThetaHat.T @ Gzz @ ThetaHat) #TODO check this 
                if (t % 50 == 0):
                    invGhxhx = np.linalg.inv(Ghxhx)
                    invGzz = np.linalg.inv(Gzz)
                else:
                    invGhxhx = sherman_morrison(invGhxhx, hxt,hxt) 
                    invGzz = sherman_morrison(invGzz, zt, zt) 
                # build estimators parameters
                ThetaHat = invGzz @ ZXt
                beta_hat = invGhxhx @ hXyt
                # build radius for ellipsoid
                logdetG += zt.T @ invGzz @ zt
                sqrt_b_2sls = calc_sqrt_b_2sls(dz, sigma_2, lamda, delta, S_hat, logdetG)
                # logs 
                ThetaHat_rec[i,t, :, :] = ThetaHat 

            # logs beta for graphs
            beta_hat_rec[i,t, :] = beta_hat 

    return regret, pred_norms, x_inv_norms, beta_hat_rec, ThetaHat_rec, norm_beta_hat__beta
```

```python
S = 1 
rhos = [1, 1.5, 2] 
sigma_1, sigma_2 = 1, 1
nb_exp, T, n_a = 20, int(5*1e3), 20 
dx = [2, 5, 8]
dz = [4, 10, 16]
dx_dz = list(zip(dx,dz))


lamda = 1e-1; lamda_IV = 1e-1
delta = 0.001
seed = 1991
algorithms = ['ridge','2sls']
labels = ['OFUL','OFUL-IV']
```

```python

def generate_data_bandit(nb_exp, T, n_a, dx, dz, theta_star, beta_star, rho, sigma_1=1, sigma_2 = 1):# first stage

    # first stage
    Z =  np.random.randn(nb_exp, T, n_a, dz) 
    # _Z = np.random.randn(T, n_a, dz) 
    # Z = np.repeat(_Z[np.newaxis, :, :, :], nb_exp, axis=0)
    _eps = sigma_1*np.random.randn(nb_exp, T, dx)
    # __eps = sigma_1*np.random.randn(T, dx)
    # _eps = np.repeat(__eps[np.newaxis, :, :], nb_exp, axis=0)
    eps = np.repeat(_eps[:, :, np.newaxis, :], n_a, axis=2)
    X = np.dot(Z, theta_star) + eps

    # second stage
    ee = 1
    _eta_tilde = np.random.randn(nb_exp, T) # TODO eliminate nb exp
    # __eta_tilde = np.random.randn(T) 
    # _eta_tilde = np.repeat(__eta_tilde[np.newaxis, :], nb_exp, axis=0)
    eta_tilde = np.repeat(_eta_tilde[:, :, np.newaxis], n_a, axis=2) 
    eta = sigma_2*(eta_tilde + rho*eps[:,:,:,:ee].sum(axis=-1))
    Y = np.dot(X, beta_star) + eta

    pmts_string = f"dx{dx}_dz{dz}_"
    # pmts_string += f"rho{rho}_"

    return Z, X, Y, pmts_string
```

```python
def get_regret_bandit(Z, X, Y,  dx, dz, theta_star, beta_star, pmts_string, a, i):
    regrets = np.zeros((nb_exp,len(algorithms),T))
    pred_norms = np.zeros((nb_exp,len(algorithms),T))
    x_inv_norms = np.zeros((nb_exp,len(algorithms),T))
    beta_hat = np.zeros((nb_exp, len(algorithms), T,dx))
    theta_hat = np.zeros((nb_exp, len(algorithms), T, dz, dx))
    beta_error = np.zeros((nb_exp, len(algorithms), T))
    pulled_indeces = np.zeros((nb_exp,T))


    regrets[:,i,:], pred_norms[:,i,:], x_inv_norms[:,i,:],\
        beta_hat[:, i, :,:] , theta_hat[:, i, :,:, :], beta_error[:,i,:]\
            = run_bandit(pmts_string, Z, X, Y, theta_star, beta_star, S, sigma_1, sigma_2, lamda, lamda_IV, delta, pulled_indeces, a, i)
    
    return regrets, pred_norms, x_inv_norms, beta_hat, theta_hat, beta_error, pulled_indeces
```

```python
def generate_data_bandit_price_sales(nb_exp, T, n_a, dx, dz, theta_star, beta_star, rho):

    # First Stage
    Z = np.random.rand(nb_exp, T, n_a, dz) 
    # _Z = np.random.rand(T, n_a, dz) 
    # Z = np.repeat(_Z[np.newaxis, :, :, :], nb_exp, axis=0)


    _eps = sigma_1*np.random.randn(nb_exp, T, dx)
    # __eps = sigma_1*np.random.randn(T, dx)
    # _eps = np.repeat(__eps[np.newaxis, :, :], nb_exp, axis=0)
    eps = np.repeat(_eps[:, :, np.newaxis, :], n_a, axis=2)

    _eta_tilde = np.random.binomial(1, 0.1, size=(nb_exp, T,1)) 
    # __eta_tilde = np.random.binomial(1, 0.01, size=(T,1)) 
    # _eta_tilde = np.repeat(__eta_tilde[np.newaxis, :], nb_exp, axis=0)
    eta_tilde = np.repeat(_eta_tilde[:, :, np.newaxis], n_a, axis=2) 

    X = np.dot(Z, theta_star) + eta_tilde + eps 

    # Second Stage
    _eta = sigma_2*np.random.randn(nb_exp, T,1)
    # __eta = sigma_2*np.random.randn(T,1)
    # _eta = np.repeat(__eta[np.newaxis, :, ], nb_exp, axis=0)
    eta = np.repeat(_eta[:, :, np.newaxis, :], n_a, axis=2)

    Y = np.dot(X, beta_star.reshape(dx,1)) + rho*eta_tilde + eta

    pmts_string = f"dx{dx}_dz{dz}_"
    return Z, X, Y, pmts_string
```

```python
def generate_data_regression(nb_exp, T, dx, dz, theta_star, beta_star, rho, sigma_1=1, sigma_2 = 1):
    # first stage
    Z =  np.random.randn(nb_exp, T, dz) 
    # _Z = np.random.randn(T, dz) 
    # Z = np.repeat(_Z[np.newaxis, :, :], nb_exp, axis=0)
    
    # _eps = sigma_1*np.random.randn(T, dx)
    # eps = np.repeat(__eps[np.newaxis, :, :], nb_exp, axis=0)
    eps = sigma_1*np.random.randn(nb_exp, T, dx)
    X = np.dot(Z, theta_star) + eps

    # second stage
    ee = 1
    eta_tilde = np.random.randn(nb_exp, T) 
    # _eta_tilde = np.random.randn(T) 
    # eta_tilde = np.repeat(_eta_tilde[np.newaxis, :], nb_exp, axis=0)
    eta = sigma_2*(eta_tilde + rho*eps[:,:,:ee].sum(axis=-1))
    Y = np.dot(X, beta_star) + eta

    pmts_string = f"dx{dx}_dz{dz}_"
    # pmts_string += f"rho{rho}_"

    return Z, X, Y, pmts_string
```

```python
def run_regression(pmts_string, Z, X, Y, theta_star, beta_star,lamda, lamda_IV, algorithm='ridge', seed=42):
    '''function used to run an instance of linear bandits using OFUL with ridge or 2SLS regression'''
    np.random.seed(np.random.randint(0,1000))
    nb_exp, T, dx = X.shape
    assert (nb_exp, T) == Z.shape[:-1]
    dz = Z.shape[-1]
    regret = np.zeros((nb_exp,T))
    norm_beta_hat__beta = np.zeros((nb_exp, T))
    beta_hat_rec = np.zeros((nb_exp,T,dx))
    ThetaHat_rec = np.zeros((nb_exp,T, dz, dx))
    
    # print(pmts_string)
    # for i in tqdm(range(nb_exp)):
    for i in range(nb_exp):
    
        
        Ghxhx, invGhxhx = lamda*np.eye(dx), np.eye(dx)/lamda
        Gxx, invGxx = lamda*np.eye(dx), np.eye(dx)/lamda
        Gzz, invGzz = lamda*np.eye(dz), np.eye(dz)/lamda # TODO is the regularizer correct here???
        ZXt =  lamda_IV*np.eye(dz, dx) # TODO np.zeros((dz, dx))
        Xyt = np.zeros(dx) 
        hXyt = np.zeros(dx)
        
        beta_hat = np.zeros(dx)
        ThetaHat = invGzz @ ZXt
        # invThetaHat = np.eye(dx)/lamda 
        invH = np.linalg.inv(ThetaHat.T @ Gzz @ ThetaHat) #np.zeros((dx,dx)) #np.ones((dx,dx))/lamda

        for t in range(T):
            zt, xt, yt = Z[i,t], X[i,t], Y[i,t]
            if t<T-1:
                xt_1 = X[i,t+1]

            if algorithm == 'vaw':
                Gxx += np.outer(xt, xt)
                if (t % 50 == 0):
                    invGxx = np.linalg.inv(Gxx)
                else:
                    invGxx = sherman_morrison(invGxx, xt,xt) 
                beta_hat = invGxx @ Xyt

            # Compute REGRET
            regret[i,t] = (np.dot(beta_hat, xt) - np.dot(beta_star, xt))**2


            # build ESTIMATORS 
            if algorithm == 'ridge':
                Xyt += xt*yt
                Gxx += np.outer(xt, xt)
                if (t % 50 == 0):
                    invGxx = np.linalg.inv(Gxx)
                else:
                    invGxx = sherman_morrison(invGxx, xt,xt) 
                beta_hat = invGxx @ Xyt
            
            if algorithm == '2sls':
                # Compute the predicted covariate
                hxt = ThetaHat.T @ zt
                # compute beta estimator
                hXyt += hxt*yt
                Ghxhx += np.outer(hxt, hxt) 
                Gzz += np.outer(zt, zt)  
                ZXt += np.outer(zt, xt)
                if (t % 50 == 0):
                    invGhxhx = np.linalg.inv(Ghxhx)
                    invGzz = np.linalg.inv(Gzz)
                else:
                    invGhxhx = sherman_morrison(invGhxhx, hxt,hxt) 
                    invGzz = sherman_morrison(invGzz, zt, zt) 
                # build estimators parameters
                ThetaHat = invGzz @ ZXt
                beta_hat = invGhxhx @ hXyt
                # logs 
                ThetaHat_rec[i,t, :, :] = ThetaHat 

            if algorithm == 'vaw':
                Xyt += xt*yt

            # logs beta for graphs
            beta_hat_rec[i,t, :] = beta_hat
            norm_beta_hat__beta[i,t] = np.linalg.norm(beta_hat-beta_star)

    return regret, beta_hat_rec, ThetaHat_rec, norm_beta_hat__beta
```

```python
def get_regret_regression(Z, X, Y,  dx, dz, theta_star, beta_star, pmts_string, a):
    regrets = np.zeros((nb_exp,T))
    beta_hat = np.zeros((nb_exp, T,dx))
    theta_hat = np.zeros((nb_exp, T, dz, dx))
    beta_error = np.zeros((nb_exp, T))


    regrets[:,:], beta_hat[:,:,:] , theta_hat[:,:,:,:], beta_error[:,:] = run_regression(pmts_string, Z, X, Y, theta_star, beta_star,lamda, lamda_IV, a)
    return regrets, beta_hat, theta_hat, beta_error
```

```python
def off_2sls(stage1, stage2, Z, X1, X2, y):
    stage1.fit(Z, X1)
    hatX1 = stage1.predict(Z)
    X = np.concatenate([hatX1, X2], axis=1)
    stage2.fit(X, y)
    r2 = stage2.score(X,y)
    haty = stage2.predict(X)
    return stage1, stage2, haty, stage1.coef_, stage2.coef_, r2

def o2sls(data, t, compare_model="linear", alpha=0.1):
    data = data.head(t)
    Z = np.array(data[["RPT", "RPN", "RPU", "RI"]]) 
    X1 = np.array(data["PG"]).reshape((-1,1))
    X2 = np.array(data["RI"]).reshape((-1,1))
    y = np.array(data["GC"]).reshape((-1,1))

    # Declare and fit comparison models
    if compare_model == 'ridge':
        stage1 = Ridge(alpha)
        stage2 = Ridge(alpha)
    elif compare_model == 'linear':
        stage1 = LinearRegression()
        stage2 = LinearRegression()

    return off_2sls(stage1, stage2, Z, X1, X2, y)
```

```python
T = len(data.index)
sp = Ridge(0.1)
betas_otsls = [] 
r2_otsls = [] 
betas_ridge = [] 
r2_ridge = [] 
for t in range(1,T+1,1):
    stage1, stage2, haty, coef1, coef2, r2 = o2sls(data, t, compare_model="ridge", alpha=0.01)
    sp.fit(np.array(data[["PG","RI"]])[:t].reshape((-1,2)), np.array(data["GC"])[:t].reshape((-1,1)))
    r2r = sp.score(np.array(data[["PG","RI"]])[:t].reshape((-1,2)), np.array(data["GC"])[:t].reshape((-1,1)))
    betas_ridge.append(sp.coef_[0][:])
    r2_ridge.append(r2r)

    betas_otsls.append(coef2[0])
    r2_otsls.append(r2)

    
r2_ridge = np.array(r2_ridge)
r2_otsls = np.array(r2_otsls)
betas_otsls = np.array(betas_otsls)
betas_ridge = np.array(betas_ridge)
```
